### Рассмотрим следующую задачу

Вы открыли свой магазин, у вас есть 5 товаров, которые вы продаете (придумайте сами название)

Вы пишете приложение, которое позволяет забронировать данные товары на складе

Из чего должно состоять ваше приложение: (10 баллов)
    
    - репозиторий на github c названием вашего проекта, сделайте его приватным 
        (Settings (строка где Code, Issues, Pull Requests) -> Manage access -> Public Repository -> Manage) (1 балл)
    - чистая структура: 
        папка lib, tests, data, один файл run.py/ app.py/ main.py, запускающий ваше приложение (1 балл)
    - все константы, названия кнопок, расположения кнопок должны храниться в конфигурационном файле 
        (пример: data.yml, data.json, data_storage.yml) - 2 баллa
    - иметь название в главном окне приложения (1 балл)
    - 5 кнопок с названиями товаров, доступных для брони в окне приложения (1 балл)
    - для каждого товара должно быть отображено количество доступных сейчас (либо внутри кнопки текст, либо блок наверху кнопок)  2 балла
    - по нажатию на кнопку с бронью товара, количество доступных должно уменьшаться (2 балла)

Нужно сделать одно (или все/ несколько) из усложнений:
    
    - вывод статистики (информации) о том, сколько товаров было забронировано - по нажатию на кнопку
      (создаем еще один объект stats = QMainWindow(), отображаем его когда нажимаем на кнопку) - 3 балла
      
    - сортировка типов товаров в порядке возрастания количества оставшихся товаров каждого типа (слева направо или снизу вверх)
        (меняем геометрию кнопок каждый раз когда нажимаем на кнопку) - 2 балла
        
    - удаление кнопки с выбором товара, если данный товар закончился
        (нужно погуглить, как использовать метод deleteLater) - 1 балл
