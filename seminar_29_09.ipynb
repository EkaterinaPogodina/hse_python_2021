### Как будет проходить собеседование?

1) вас могут попросить немного рассказать о себе (тут стоит сказать где выучитесь, что любите программировать, назвать любимый язык программирование, что больше всего привлекает) - это не провакационный вопрос, эту технику часто используют чтобы наладить контакт с собеседником

2) вам расскажут условие задачи - нужно внимательно его выслушать и задать уточняющиее вопросы, далее план примерно следующий:

    - нужно придумать решение, проговорить его в слух, оценить его сложность (об оценке сложности поговорим в следующем модуле) 
    - если решение будет неоптимальное, собеседущий попросит вас подумать еще, возможно даст подсказку - если что, не расстраивайтесь подсказкам - это нормально
    
    - далее нужно начать писать решение:
        - решение должно быть рабочим (псевдокод это плохо)
        - избегайте громоздких конструкций и копирование кода (сделайте отдельную функцию), они же дадут вам себя запутать
        - посторайтесь не допускать багов, в голове прогоните все краевые случаи - можете рассказать собеседующему о том, что проверили вот на таких-то примерах и решение рабочее
        - если не помните, какую-то языковую конструкцию (например, как писать исключения, как привести строку к нижнему регистру) - честно скажите об этом
        
    - на что будет смотреть собеседующий:
        - на скорость (чем больше кода писал в жизни - тем быстрее переведешь алгоритм в код) и чистоту написания кода
        - смогли ли вы довести решения до конца без багов
        - насколько хорошо вы знаете свой любимый язык

3) после того как задача будет решена, вам дадут следующую задачу, стандартное собеседование - это 2 задачи: первая потенциально пишется за 15 минут, вторая - за пол часа

4) вам предложат задать вопросы (можно спросить у собеседующего чем он занимается, можно сказать что пока вопросов нет)
